**Zad. 4**. Dla \begin{align}
        A = \begin{pmatrix}
        4 & 5 & 2 \\
        -6 & -1 & 3\\
        2 & -2 & 5
        \end{pmatrix}
    \end{align}
    oraz
    \begin{align}
    B = \begin{pmatrix}
        -2 & 6 & -2 \\
        1 & -3 & -3\\
        4 & 8 & -1
        \end{pmatrix}
    \end{align}
    sprawdź, czy zachodzi:

*   $A^T + B^T = (A + B)^T$
*   $AB = BA$
*   $(AB)^T = A^T B^T$
*   $(AB)^T = B^T A^T$,
gdzie `T` oznacza transpozycję macierzy.
